```python
import pandas as pd
import numpy as np
from konlpy.tag import *
```

```python
metadata = pd.read_csv('data_nountag_all_song.csv', encoding='cp949')
```

```python
metadata.head()
```

```python
synop = metadata['synop']
```

```python
# 개행문자 제거 후 synop_list에 넣음

synoplist = []
for i in synop:
    synops = i.replace('\n', '')
    synoplist.append(synops)    
```

```python
synoplist[1]
```

## 형태소 분석

```python
han = Hannanum()

#불용어 제거 안함.
han.nouns(synoplist[0])
```

```python
n_synoplist = []

for synop in synoplist:
    s_nouns = ' '.join(han.nouns(synop))
    #print(s_nouns)
    n_synoplist.append(s_nouns)
```

```python
metadata.head()
```

```python
n_synop = n_synoplist
```

## 새 데이터프레임, csv파일 저장

```python
metadata['n_synop'] = n_synop
```

```python
metadata.drop(['synop'], 1)
```

```python
metadata.to_csv('data_all_nsynop_dayoung.csv', encoding='utf-8')
```
