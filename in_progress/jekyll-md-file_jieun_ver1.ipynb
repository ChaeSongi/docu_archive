```python
route = "C:\\Users\\USER\dataitgirls\documen\_posts"
file_path = route + '\\' + df['dates'][0] + '-' + '0' + '.md'
f = codecs.open(file_path,'w','utf-8')

template = """---
layout: post
title: {{ epi }}
date: {{ dates }}
image: 'http://res.cloudinary.com/dm7h7e8xj/image/upload/c_fill,h_399,w_760/v1503153729/golpe_ghb84o.jpg'
description: {{ title }}
category: {{ big_tag }}
tags: {{ tags }}
---
{{ contents }}
"""

tags = df['tags'][0].split(',')
tags = [ tag.strip() for tag in tags ]
all_tags = []
for tag in tags:
    all_tags.append(tag)
temp = "\n- ".join(all_tags)
content = "\n- {}".format(temp)
content

template = template.replace('{{ epi }}', df['epi'][0])
template = template.replace('{{ dates }}', df['dates'][0])
template = template.replace('{{ title }}', df['title'][0])
template = template.replace('{{ big_tag }}', df['tv'][0])
template = template.replace('{{ tags }}', content)
template = template.replace('{{ contents }}', df['synop'][0])


f.write(template)
f.close()
```
