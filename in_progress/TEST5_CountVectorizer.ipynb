```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(min_df=1)
#count_matrix = count.fit_transform(metadata)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
#metadata = metadata.reset_index()
indices = pd.Series(metadata.index, index=metadata['epi'])
```

```python
def get_recommendations(episode, cosine_sim2=cosine_sim2):
    #제목과 일치하는 에피소드 인덱스 가지고옴
    idx = indices[epi]
    
    sim_scores = list(enumerate(cosine_sim2[idx]))
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    #비슷한 에피소드 상위 10개
    sim_scores = sim_scores[1:6]
    
    # indices
    epi_indices = [i[0] for i in sim_scores]
    
    return metadata['epi'].iloc[epi_indices]
```

```python
get_recommendations('비밀 땅 파미르 1부 세계 지붕', cosine_sim2)
```
