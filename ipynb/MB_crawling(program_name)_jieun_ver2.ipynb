# MBC 프로그램 회차, 방송날짜, 시놉시스

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# 프로그램 링크 수집
driver.get('http://vodmall.imbc.com/genre/genre_sub.aspx?genre=13')
pages = range(0, 6)
all_links = []
all_titles = []
for page in pages:
    driver.find_elements_by_css_selector('.masterPaging ul li a')[page].click()
    WebDriverWait(driver, 4)
    links = driver.find_elements_by_css_selector('#ImageTable > li > ul > li:first-child > a')
    titles = driver.find_elements_by_css_selector('#ImageTable > li > ul > li:first-child > a > span')
    all_titles += [title.text for title in titles]
    all_links += [link.get_attribute('href') for link in links]
```

```python
# 프로그램 페이지에 접속하여 회차 ID 번호 수집하기
episodes = []
titles = []
episode_keys = []
# all_links[:3]
for link, title in zip(all_links, all_titles):
    driver.get(link)
    options = driver.find_elements_by_css_selector('#ddlEpisode option')
    for option in options:
        episode_keys.append(option.get_attribute('value'))
        episodes.append(option.text)
        titles.append(title)
        print(title)

# result => 0 1000834100083100000
```

```python
import requests
all_synopsis = []

url_prefix = 'http://vodmall.imbc.com/util/vodUtil.aspx?kind=image&pagenum=1&pagesize=5&cornerFlag=0&ContentTypeID=1&ProgramGroupID=0&progCode='    

# episode_keys[:3]
for key in episode_keys:
    res = requests.get(url_prefix + key)
    all_synopsis.append(res.text)
```

```python
episode_url = []
ep_url_prefix = 'http://vodmall.imbc.com/genre/genre_program.aspx?progCode='
for key in episode_keys:
    episode_url.append(ep_url_prefix+key)
```
